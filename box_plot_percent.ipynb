```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./csv/wgan_new_result_match.csv', sep='\t', header=None)
df_t = df.T
df_t.columns = df_t.iloc[0]
df_t /= 10
df = df_t[1:]
boxplot = df.boxplot()
boxplot.set_xlabel('wavelength (nm)')
boxplot.set_ylabel('occurrences')
boxplot.set_ylim([0, 1])
plt.savefig("./figures/wgan_new_result_match_occurrences.png", format="png", dpi=600)
```

```python
df=pd.read_csv('./csv/wgan_new_result_truth.csv', sep='\t', header=None)
df_t = df.T
df_t.columns = df_t.iloc[0]
df_t /= 120
df = df_t[1:]
boxplot = df.boxplot()
boxplot.set_xlabel('wavelength (nm)')
boxplot.set_ylabel('occurrences')
boxplot.set_ylim([0, 0.4])
plt.savefig("./figures/wgan_new_result_truth_occurrences.png", format="png", dpi=600)

```
